```python
def  item_fat(weight):
    if (weight <= 5):
        return("low fat")
    elif((weight >=9) & (weight <=11 )):
        return("Regular")
    elif((weight>= 15) & (weight<= 17)):
        return("LF")
    else:
        return("reg")

```
